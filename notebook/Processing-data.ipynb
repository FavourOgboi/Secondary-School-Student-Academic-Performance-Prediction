# Schools Student Performance Data Set

### Source:
This dataset was collected through field visits to secondary schools in the Ikorodu area of Lagos State, Nigeria. Schools visited include:
- Elepe Community Senior High School
- Tiakod College
- Kingsfields Group of Schools

To gather the data, students were given a **structured form** to fill out, providing personal, academic, and socio-economic details. The responses were later compiled and entered into an Excel spreadsheet for analysis. 


The dataset aims to analyze students' performance and their academic environment to better understand the factors influencing their educational outcomes. The students included are from Senior Secondary School 1 (SS1), Senior Secondary School 2 (SS2) and Senior Secondary School 2 (SS3).

---

### Data Set Information:
This dataset contains detailed information about students, focusing on their academic performance, living situation, parental involvement, and personal experiences in school. The data was collected through interviews with school students after seeking for permission from principal and granted access.

The primary goal of this data set is to assess how various social, economic, and psychological factors influence students' academic outcomes and their future educational plans.

---

### Attribute Information:
- **Student ID**: A unique identifier assigned to each student.
- **Gender**: The gender of the student (Male/Female).
- **Age Group**: The age category of the student (e.g., 14-15, 16-17).
- **Living Situation**: Describes the student’s living situation (e.g., Living with Parents, Guardian, Hostel).
- **Parent Education Level**: The highest education level attained by the student’s parents (e.g., No Formal Education, Primary, Secondary, Tertiary).
- **Family Financial Status**: The economic status of the student's family (e.g., Low Income, Middle Income, High Income).
- **Class**: The student's academic class (SS1 or SS2).
- **Department**: The student's academic department (Science, Humanities, Commercial).
- **Performance in English**: The student's score in English (0-100).
- **Performance in Maths**: The student's score in Mathematics (0-100).
- **Performance in Biology**: The student's score in Biology (0-100) (Science students only).
- **Performance in Physics**: The student's score in Physics (0-100) (Science students only).
- **Performance in Chemistry**: The student's score in Chemistry (0-100) (Science students only).
- **Performance in Lit in English**: The student's score in Literature (0-100) (Humanities students only).
- **Performance in Government**: The student's score in Government (0-100) (Humanities students only).
- **Performance in CRS**: The student's score in Christian Religious Studies (0-100) (Humanities students only).
- **Performance in Commerce**: The student's score in Commerce (0-100) (Commercial students only).
- **Performance in Accounting**: The student's score in Accounting (0-100) (Commercial students only).
- **Performance in Economics**: The student's score in Economics (0-100) (Commercial students only).
- **Study Hours per Week**: The average number of hours the student dedicates to studying each week.
- **Extra Tutoring**: Whether the student receives additional tutoring outside regular school hours (Yes/No).
- **School Attendance**: The student's attendance rate in percentage.
- **Extracurricular Participation**: Indicates if the student participates in extracurricular activities (Yes/No).
- **Considered Dropping Out**: Whether the student has considered dropping out of school (Yes/No).
- **Reason To Dropout**: The main reason given for considering dropping out (e.g., Financial Issues, Academic Struggles, Family Problems).
- **Experienced Bullying**: Whether the student has experienced bullying in school (Yes/No).
- **Type of Bullying**: The type of bullying experienced (e.g., Physical, Verbal, Cyber).
- **Comfort in School Environment**: The student's level of comfort in the school environment (Comfortable, Neutral, Uncomfortable).
- **Trusted Teacher**: Whether the student has a teacher they trust (Yes/No).
- **Peer Pressure**: Whether the student feels they experience peer pressure in school (Yes/No).
- **Parents Help with Homework**: Whether the student's parents assist with homework (Yes/No).
- **Parents Attend Parent-Teacher Meeting**: Whether the student's parents attend parent-teacher meetings (Yes/No).
- **Parents Encourage Higher Education**: Whether the student's parents encourage them to pursue higher education (Yes/No).
- **Stress About Schoolwork**: The student's stress level concerning schoolwork (Low, Moderate, High).
- **Confidence in Academic Ability**: The student's confidence in their academic ability (Low, Average, High).
- **Access to Counseling**: Whether the student has access to counseling services at school (Yes/No).
- **Home Learning Resources**: The availability of learning resources at home (e.g., Internet, Books, Computer).
- **School Learning Resources**: The availability of learning resources in school (e.g., Library, Computer Lab).
- **Motivation for Attending School**: The student's primary motivation for attending school (e.g., Desire to Learn, Parent's Expectation, Peer Influence).
- **Plan to Attend University**: Whether the student plans to attend university after secondary school (Yes/No).
- **JAMB Goal Score**: The student’s goal score for the Joint Admissions and Matriculation Board (JAMB) exam.
- **Extracurricular Hours**: The average number of hours the student spends on extracurricular activities per week.
- **Free Time Activities**: The student's primary free time activities (e.g., Sports, Reading, Social Media).
- **Language at Home**: The language(s) spoken at home (e.g., English, Yoruba, Pidgin).
- **Involved in Work Outside School**: Whether the student is involved in work outside of school (Yes/No).
- **Last Exam Performance**: The student's overall performance in their last exam (e.g., Passed, Failed, Excellent).
- **Foundational Knowledge**: The student’s foundational knowledge in their department (rated as Strong, Average, or Weak).

```python
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")

df = df.drop_duplicates()
```

```python
# Converting all object columns (except 'Student ID') to category
categorical_columns = df.columns.difference(['Student ID'])

df[categorical_columns] = df[categorical_columns].astype('category')

print(df.dtypes)
```

```python
# Checking for the unique values for each categorical column
for col in categorical_columns:
    print(f'Unique values in {col}: {df[col].unique()}')
```

### **Checking distributions in the dataset for proper cleaning**
***

Conducting an initial distribution analysis of the dataset. This analysis revealed several discrepancies in some of the columns, which will be addressed to ensure the dataset is clean and ready for `further analysis`.

```python
# List of columns to skip
columns_to_skip = ['Free Time Activities', 'Home Learning Resources', 'School Learning Resources','Type of Bullying']  # Add other column names as needed

# Assuming categorical_columns is defined elsewhere
for col in categorical_columns:
    if col not in columns_to_skip:  # Only plot if the column is not in the skip list
        sns.countplot(data=df, x=col)
        plt.title(f'Distribution of {col}')
        plt.xticks(rotation=45)
        plt.show()
```

##### Cleaning the "Age Group" column to standardize inconsistent entries like `15 -16` and `17 -18` for accurate analysis.

```python
# Cleaning the Age Group column using cat.rename_categories to avoid warnings
df['Age Group'] = df['Age Group'].cat.rename_categories({
    '15 -16': '15 - 16',
    '17 -18': '17 - 18'
})

print("Unique values in Age Group after cleaning:", df['Age Group'].unique())
```

##### Cleaning the "Comfort in School Environment" column to standardize inconsistent entries like `Very Comfortable` and `Comfortable` for accurate analysis.

```python
# Renaming the categories directly using rename_categories
df['Comfort in School Environment'] = df['Comfort in School Environment'].cat.rename_categories({
    'Very Comfortable': 'Comfortable'
})

print("Unique values in Comfort in School Environment after cleaning:", df['Comfort in School Environment'].unique())
```

##### Cleaning the "Confidence in Academic Ability" column to standardize inconsistent entries like `Not very confident` and `Confidence in Academic Ability` for accurate analysis.

```python
# Replacing the inconsistent entries
df['Confidence in Academic Ability'] = df['Confidence in Academic Ability'].replace({
    'Not very confident': 'Not confident',
    'Confidence in Academic Ability': 'Very confident'
})

df['Confidence in Academic Ability'] = df['Confidence in Academic Ability'].cat.set_categories([
    'Not confident', 
    'Somewhat confident', 
    'Very confident'
])

print("Unique values in Confidence in Academic Ability after cleaning:", df['Confidence in Academic Ability'].unique())
```

##### Cleaning the "Considered Dropping Out" column to standardize inconsistent entries like `Yes` and `Often` for accurate analysis.

```python
# Replacing 'Yes' with 'Often'
df['Considered Dropping Out'] = df['Considered Dropping Out'].replace('Yes', 'Often')

df['Considered Dropping Out'] = df['Considered Dropping Out'].cat.set_categories([
    'Never', 
    'Sometimes', 
    'Often'
])

print("Unique values in Considered Dropping Out after cleaning:", df['Considered Dropping Out'].unique())
```

##### Cleaning the "Extra Tutoring" column to standardize inconsistent entries like `Yes` and `Regularly` for accurate analysis.

```python
# Replacing 'Yes' with 'Regularly'
df['Extra Tutoring'] = df['Extra Tutoring'].replace('Yes', 'Regularly')

df['Extra Tutoring'] = df['Extra Tutoring'].cat.set_categories([
    'No', 
    'Occasionally', 
    'Regularly'
])

print("Unique values in Extra Tutoring after cleaning:", df['Extra Tutoring'].unique())
```

- Checking the foundational knowledge and previous performance of students who have "Extracurricular Hours" as 10 - 15 hours

```python
# Filtering the DataFrame for students with 'Extracurricular Hours' as '10 - 15 hours'
filtered_students = df[df['Extracurricular Hours'] == '10 - 15 hours'].count()

foundational_knowledge = filtered_students['Foundational Knowledge']
previous_performance = filtered_students['Last Exam Performance']

print("Foundational Knowledge of students with 10 - 15 hours of Extracurricular Activities:")
print(foundational_knowledge)

print("\nPrevious Performance of students with 10 - 15 hours of Extracurricular Activities:")
print(previous_performance)
```

> Just 3 students.

##### Cleaning the "Extracurricular Hours" column to standardize inconsistent entries like `10 - 15 hours` to `More than 10 hours`

```python
# Replacing '10 - 15 hours' with 'More than 10 hours'
df['Extracurricular Hours'] = df['Extracurricular Hours'].replace('10 - 15 hours', 'More than 10 hours')

df['Extracurricular Hours'] = df['Extracurricular Hours'].cat.set_categories([
    'Less than 5 hours', 
    '5 - 10 hours', 
    'More than 10 hours'
])

print("Unique values in Extracurricular Hours after cleaning:", df['Extracurricular Hours'].unique())
```

##### Cleaning the "Family Financial Status" column to standardize inconsistent entries and remove duplicates.
- Unifying the two entries `We struggle to meet basic needs` into a single, consistent entry.

```python
# Replacing the inconsistent entries
df['Family Financial Status'] = df['Family Financial Status'].replace(
    {
        'We Struggle to meet basic needs': 'We struggle to meet basic needs'
    }
)

df['Family Financial Status'] = df['Family Financial Status'].cat.set_categories([
    'We meet our needs but can\'t afford luxuries',
    'We can afford some luxuries',
    'We are financially comfortable',
    'We struggle to meet basic needs'
])

print("Unique values in Family Financial Status after cleaning:", df['Family Financial Status'].unique())
```

##### Standardizing the "Foundational Knowledge" column by merging inconsistent entries to improve data accuracy.

```python
# Cleaning the Foundational Knowledge column by standardizing entries
df['Foundational Knowledge'] = df['Foundational Knowledge'].str.title().replace('Strong', 'Strong')

print("Unique values in Foundational Knowledge after cleaning:", df['Foundational Knowledge'].unique())
```

##### Standardizing the "Involved in Work Outside School" column to ensure consistent responses for accurate analysis.
- Unifying the two entries `Yes` and `Frequently` into a single response.

```python
# Cleaning the Involved in Work Outside School column by merging responses
df['Involved in Work Outside School'] = df['Involved in Work Outside School'].replace('Yes', 'Frequently')

print("Unique values in Involved in Work Outside School after cleaning:", df['Involved in Work Outside School'].unique())
```

##### Standardizing the "JAMB Goal Score" column to eliminate duplicate entries for accurate analysis.
- Merging the entries `300 And Above` and `300 and Above` into a single response

```python
# Cleaning the JAMB Goal Score column by merging duplicate entries
df['JAMB Goal Score'] = df['JAMB Goal Score'].replace('300 and Above', '300 And Above')

print("Unique values in JAMB Goal Score after cleaning:", df['JAMB Goal Score'].unique())
```

##### Standardizing the "Last Exam Performance" column to unify inconsistent entries for accurate analysis.
- Change `60% -69%` to `60% - 69%`.
- Merge `40% - 49%` into `40% - 59%`.

```python
# Cleaning the Last Exam Performance column by correcting duplicate entries
df['Last Exam Performance'] = df['Last Exam Performance'].replace({
    '60% -69%': '60% - 69%',  # Correcting the formatting
    '40% - 49%': '40% - 59%'   # Merging with the broader category
})

# Check the unique values after cleaning
print("Unique values in Last Exam Performance after cleaning:", df['Last Exam Performance'].unique())
```

##### Renaming the "Perfomance in Commerce" column to "Performance in Commerce" for correct spelling and standardizing inconsistent entries like `Nil` and `NIl` for accurate analysis.


```python
# Rename the column for correct spelling
df.rename(columns={'Perfomance in Commerce': 'Performance in Commerce'}, inplace=True)
```

```python
# Standardizing the entries by replacing 'NIl' with 'Nil'
df['Performance in Commerce'] = df['Performance in Commerce'].replace('NIl', 'Nil')

print("Unique values in Performance in Commerce after cleaning:", df['Performance in Commerce'].unique())
```

##### Cleaning the `Performance in Government` column to standardize inconsistent entries, specifically replacing `M` with `Nil` for accurate analysis.

##### Cleaning the `Plan to Attend University` column to standardize inconsistent entries by replacing variations of `Yes` with a single standardized entry for accurate analysis.

```python
# Standardizing the entries in 'Plan to Attend University' column
df['Plan to Attend University'] = df['Plan to Attend University'].replace({'yes': 'Yes', "I'm not sure": 'Not Sure'})

unique_values = df['Plan to Attend University'].unique()
print("Unique values in Plan to Attend University after cleaning:", unique_values)
```

##### Cleaning the `Stress About Schoolwork1 column to standardize inconsistent entries, merging `Rearly` to `Rarely` and ensuring a consistent set of responses for accurate analysis.

```python
df['Stress About Schoolwork'] = df['Stress About Schoolwork'].replace({
    'Rearly': 'Rarely',  # Correcting the spelling of 'Rearly' to 'Rarely'
    'Somewhat stressed': 'Sometimes'  # Standardizing 'Somewhat stressed' to 'Sometimes'
})

unique_values = df['Stress About Schoolwork'].unique()
print("Unique values in Stress About Schoolwork after cleaning:", unique_values)
```

##### Cleaning the `Study Hours per Week` column to consolidate overlapping categories and ensure consistency for accurate analysis.
