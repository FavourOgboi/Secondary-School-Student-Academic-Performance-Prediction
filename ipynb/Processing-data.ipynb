# Schools Student Performance Data Set

### Source:
This dataset was collected through field visits to secondary schools in the Ikorodu area of Lagos State, Nigeria. Schools visited include:
- Elepe Community Senior High School
- Tiakod College
- Kingsfields Group of Schools

To gather the data, students were given a **structured form** to fill out, providing personal, academic, and socio-economic details. The responses were later compiled and entered into an Excel spreadsheet for analysis. 


The dataset aims to analyze students' performance and their academic environment to better understand the factors influencing their educational outcomes. The students included are from Senior Secondary School 1 (SS1), Senior Secondary School 2 (SS2) and Senior Secondary School 2 (SS3).

---

### Data Set Information:
This dataset contains detailed information about students, focusing on their academic performance, living situation, parental involvement, and personal experiences in school. The data was collected through interviews with school students after seeking for permission from principal and granted access.

The primary goal of this data set is to assess how various social, economic, and psychological factors influence students' academic outcomes and their future educational plans.

---

### Attribute Information:
- **Student ID**: A unique identifier assigned to each student.
- **Gender**: The gender of the student (Male/Female).
- **Age Group**: The age category of the student (e.g., 14-15, 16-17).
- **Living Situation**: Describes the student’s living situation (e.g., Living with Parents, Guardian, Hostel).
- **Parent Education Level**: The highest education level attained by the student’s parents (e.g., No Formal Education, Primary, Secondary, Tertiary).
- **Family Financial Status**: The economic status of the student's family (e.g., Low Income, Middle Income, High Income).
- **Class**: The student's academic class (SS1 or SS2).
- **Department**: The student's academic department (Science, Humanities, Commercial).
- **Performance in English**: The student's score in English (0-100).
- **Performance in Maths**: The student's score in Mathematics (0-100).
- **Performance in Biology**: The student's score in Biology (0-100) (Science students only).
- **Performance in Physics**: The student's score in Physics (0-100) (Science students only).
- **Performance in Chemistry**: The student's score in Chemistry (0-100) (Science students only).
- **Performance in Lit in English**: The student's score in Literature (0-100) (Humanities students only).
- **Performance in Government**: The student's score in Government (0-100) (Humanities students only).
- **Performance in CRS**: The student's score in Christian Religious Studies (0-100) (Humanities students only).
- **Performance in Commerce**: The student's score in Commerce (0-100) (Commercial students only).
- **Performance in Accounting**: The student's score in Accounting (0-100) (Commercial students only).
- **Performance in Economics**: The student's score in Economics (0-100) (Commercial students only).
- **Study Hours per Week**: The average number of hours the student dedicates to studying each week.
- **Extra Tutoring**: Whether the student receives additional tutoring outside regular school hours (Yes/No).
- **School Attendance**: The student's attendance rate in percentage.
- **Extracurricular Participation**: Indicates if the student participates in extracurricular activities (Yes/No).
- **Considered Dropping Out**: Whether the student has considered dropping out of school (Yes/No).
- **Reason To Dropout**: The main reason given for considering dropping out (e.g., Financial Issues, Academic Struggles, Family Problems).
- **Experienced Bullying**: Whether the student has experienced bullying in school (Yes/No).
- **Type of Bullying**: The type of bullying experienced (e.g., Physical, Verbal, Cyber).
- **Comfort in School Environment**: The student's level of comfort in the school environment (Comfortable, Neutral, Uncomfortable).
- **Trusted Teacher**: Whether the student has a teacher they trust (Yes/No).
- **Peer Pressure**: Whether the student feels they experience peer pressure in school (Yes/No).
- **Parents Help with Homework**: Whether the student's parents assist with homework (Yes/No).
- **Parents Attend Parent-Teacher Meeting**: Whether the student's parents attend parent-teacher meetings (Yes/No).
- **Parents Encourage Higher Education**: Whether the student's parents encourage them to pursue higher education (Yes/No).
- **Stress About Schoolwork**: The student's stress level concerning schoolwork (Low, Moderate, High).
- **Confidence in Academic Ability**: The student's confidence in their academic ability (Low, Average, High).
- **Access to Counseling**: Whether the student has access to counseling services at school (Yes/No).
- **Home Learning Resources**: The availability of learning resources at home (e.g., Internet, Books, Computer).
- **School Learning Resources**: The availability of learning resources in school (e.g., Library, Computer Lab).
- **Motivation for Attending School**: The student's primary motivation for attending school (e.g., Desire to Learn, Parent's Expectation, Peer Influence).
- **Plan to Attend University**: Whether the student plans to attend university after secondary school (Yes/No).
- **JAMB Goal Score**: The student’s goal score for the Joint Admissions and Matriculation Board (JAMB) exam.
- **Extracurricular Hours**: The average number of hours the student spends on extracurricular activities per week.
- **Free Time Activities**: The student's primary free time activities (e.g., Sports, Reading, Social Media).
- **Language at Home**: The language(s) spoken at home (e.g., English, Yoruba, Pidgin).
- **Involved in Work Outside School**: Whether the student is involved in work outside of school (Yes/No).
- **Last Exam Performance**: The student's overall performance in their last exam (e.g., Passed, Failed, Excellent).
- **Foundational Knowledge**: The student’s foundational knowledge in their department (rated as Strong, Average, or Weak).

```python
# Checking for missing values in the dataset
missing_values = df.isnull().sum()

missing_values = missing_values[missing_values > 0]

# Display the columns with missing values and the count
print(missing_values)
```

No missing values

`NB`: I have to remove white spaces from column names and each entries because python is very sensitive to this.

```python
# Removing whitespace from column names
df.columns = df.columns.str.strip()

# Removing whitespace from all entries in the DataFrame
for col in df.columns:
    if df[col].dtype == 'object':  # Checking if the column is of string type is object(text) then strip
        df[col] = df[col].str.strip()
        
df.head()
```

CHECKING FOR DUPLICATES

```python
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")

df = df.drop_duplicates()
```

```python
# Converting all object columns (except 'Student ID') to category
categorical_columns = df.columns.difference(['Student ID'])

df[categorical_columns] = df[categorical_columns].astype('category')

print(df.dtypes)
```

```python
# Checking for the unique values for each categorical column
for col in categorical_columns:
    print(f'Unique values in {col}: {df[col].unique()}')
```

Checking distributions in the dataset for proper cleaning

```python
for col in categorical_columns:
    sns.countplot(data=df, x=col)
    plt.title(f'Distribution of {col}')
    plt.xticks(rotation=45)
    plt.show()
```
